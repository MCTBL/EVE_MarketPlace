```python
def getByGroupId(id):
    url = "https://esi.evepc.163.com/latest/markets/groups/" + str(id) + "/?datasource=serenity&language=zh"
    return re.get(url).json()

def getByTypesId(id):
    url = "https://esi.evepc.163.com/latest/universe/types/" + str(id) + "/?datasource=serenity&language=zh"
    dic = re.get(url).json()
    if "dogma_attributes" in dic:
        dic.pop("dogma_attributes")
    if "dogma_effects" in dic:
        dic.pop("dogma_effects")
    return dic

def getLatestHistoryById(id):
    url = "https://esi.evepc.163.com/latest/markets/10000002/history/?datasource=serenity&type_id=" + str(id)
    dic = re.get(url).json()
    return dic[-1]
```
